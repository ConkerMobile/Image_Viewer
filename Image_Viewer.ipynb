```python
from tkinter import *
from tkinter import filedialog


root = Tk()
def askfile():
    filename = filedialog.askopenfile(filetypes=[("JPEG", "*.jpg"), ("PNG", "*.png")]).name
    from PIL import Image, ImageTk

    class Window(Frame):
        def __init__(self, master=None):
            Frame.__init__(self, master)
            self.master = master
            self.pack(fill=BOTH, expand=1)
            
            load = Image.open(filename)
            w, h = load.size
            root.geometry(f"{w}x{h}")
            render = ImageTk.PhotoImage(load)
            img = Label(self, image=render)
            img.image = render
            img.place(x=0, y=0)

            
    app = Window(root)
    root.mainloop()

openfile = Button(root, text="Open File", fg='white', bg='green', command=askfile)
close = Button(root, text="Close", fg='white', bg='red', command=root.destroy)
close.pack()
openfile.pack()
root.mainloop()
```

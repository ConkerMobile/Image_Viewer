```python
from tkinter import *
from tkinter import filedialog


root = Tk()
img = Label(root)
def askfile():
    global img
    img.configure(image='')
    filename = filedialog.askopenfile(filetypes=[("JPEG", "*.jpg"), ("PNG", "*.png")]).name
    from PIL import Image, ImageTk

    load = Image.open(filename)
    w, h = load.size
    root.geometry(f"{w+70}x{h+70}")
    render = ImageTk.PhotoImage(load)
    img = Label(root, image=render)
    img.image = render
    img.place(x=50, y=50)
    root.mainloop()

openfile = Button(root, text="Open File", fg='white', bg='green', command=askfile)
close = Button(root, text="Close", fg='white', bg='red', command=root.destroy)
close.pack()
openfile.pack()
root.mainloop()
```
